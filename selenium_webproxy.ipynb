```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import pandas as pd
import yaml
import os
import requests
```

```python
response = requests.get("https://free-proxy-list.net/")
proxy_list = pd.read_html(response.text)[0]
proxy_list["url"] = "https://" + proxy_list["IP Address"] + ":" + proxy_list["Port"].astype(str)
https_proxies = proxy_list["url"]
https_proxies.count()
```

```python
with open("headers.yml") as f_headers:
    browser_headers = yaml.safe_load(f_headers)
```

```python
url = "https://httpbin.org/ip"
good_proxies = set()
headers = browser_headers["Chrome"]
```

```python
for proxy_url in https_proxies:
    proxies = {
        "http": proxy_url,
        "https": proxy_url
    }
    print(proxy_url)
    try:
        response = requests.get(url, headers=headers, proxies=proxies, timeout=1)
        good_proxies.add(proxy_url)
        print(f"Proxy {proxy_url} OK, added to good_proxy list")
    except Exception:
        pass
len(good_proxies)
```

```python
for proxy_url in good_proxies:
    proxy = proxy_url.replace("http://", "")

    firefox_capabilities = webdriver.DesiredCapabilities.FIREFOX
    firefox_capabilities["marionette"]= True

    options = Options()
    options.add_argument('--proxy-server=%s' % proxy)
    os.environ['PATH'] = "Users\ASUS\PycharmProjects\geckodriver.exe" + os.environ['PATH']
    driver = webdriver.Firefox(options=options)
    try:
        driver.get("https://httpbin.org/ip")
    except Exception:
        pass
```
